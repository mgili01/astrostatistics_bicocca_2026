General sigma

```python
sigma = 2
```

```python
def fun(x):
    return (sigma*np.sqrt(2*np.pi) * x**3)/2
```

```python
xi = np.absolute(np.random.normal(0,sigma,100000))
```

```python
integral = np.mean(fun(xi))
```

```python
integral
```
