```python
import numpy as np
import scipy.stats
import pylab as plt
plt.rcParams['figure.figsize'] = [8, 8]
```

```python
deaths = [0, 1, 2, 3, 4]
groups = [109, 65, 22, 3, 1]
```

```python
N = np.sum(groups)
```

```python
prob = groups / N
```

```python
plt.bar(deaths, prob)

plt.xlabel("Deaths per corps")
plt.ylabel("Probability")

plt.show()
```

```python
lamb = np.sum(np.array(deaths)*np.array(groups))/N
print(lamb)
```

```python
from math import factorial
prob_poisson = [(lamb**i * np.exp(-lamb))/ factorial(i) for i in deaths]
```

```python
plt.bar(deaths-0.15, prob, width=0.3, label="Data")
plt.bar(deaths+0.15, prob_poisson, width=0.3, label="Poisson Fit")

plt.xlabel("Number of deaths")
plt.ylabel("Probability")
```
